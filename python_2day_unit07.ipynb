```python
import csv

f =  open('age.csv')
data = csv.reader(f)
result = []

area = input('인구구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 :')

for row in data :
#    if '신도림' in row[0]:
    if area in row[0]:
        for i in row[3:]:
            result.append(int(i))

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```
